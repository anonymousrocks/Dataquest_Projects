```python
import pandas as pd 
import matplotlib.pyplot as plt 
%matplotlib inline 
```

```python
recent_grads = pd.read_csv("recent-grads.csv")
recent_grads.iloc[0]
```

```python
recent_grads = recent_grads.dropna() 
```

```python
cleaned_data_count = len(recent_grads)
cleaned_data_count
```

```python
fig, axes = plt.subplots(nrows= 3, ncols = 2, figsize = (10,20))
ax1 = recent_grads.plot(ax = axes[0,0], x="Sample_size", y = "Median", kind = "scatter",
                        rot = 90, title = "Sample Size vs Salary")
ax2 = recent_grads.plot(ax = axes[0,1], x = "Sample_size", y= "Unemployment_rate", kind = "scatter"
                       , rot = 90, title = "Sample Size vs Unemployment Rate")
ax3 = recent_grads.plot(ax = axes[1,0], x= "Full_time", y = "Median", kind = "scatter",
                        rot = 90 , title = " Full time employess (n) vs Median Salary")
ax4= recent_grads.plot( ax = axes[1,1], x= "ShareWomen", y="Unemployment_rate", kind = "scatter", 
                       rot = 90, title = "Women Proportion vs Unemployment Rate")
ax5 = recent_grads.plot( ax = axes[2,0], x = "Men", y="Median", kind = "scatter", 
                       rot = 90, title = "Men vs Salary")
ax6 = recent_grads.plot( ax = axes[2,1], x="Women", y= "Median", kind = "scatter",
                       rot = 90, title = "Women vs Salary")
plt.show()
```

### Do Students in more popular majors make more money ? 

```python
ax1 = recent_grads.plot(x="Total", y = "Median", kind = "scatter", rot = 90)
ax1.set_title("Total Student per Major vs Salary")
```

This result shows that it has no feasible correlation between total student per major with salary 

### Do Students that majored in subjects that were majority female make more money? 

```python
ax2= recent_grads.plot( x= "ShareWomen", y="Median", kind = "scatter", 
                       rot = 90, title = "Women Proportion vs Median")
```

This have a very small correlation as the increase in share of women earn less money 

### Is there any link between the number of full-time employees and median salary?

```python
ax3 = recent_grads.plot( x= "Full_time", y = "Median", kind = "scatter", rot = 90 , title = " Full time employess (n) vs Median Salary")
```

Based on this scatter plot, it is found out that there is not correlation between number of full time employess and median salary

```python
fig = plt.figure(figsize = (10,10))
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)
recent_grads[:10].plot.bar(ax = ax1 , x = "Major", y = "ShareWomen", title = "First 10 Women")
ax2 = recent_grads[-10:].plot.bar(ax = ax2 , x = "Major", y ="ShareWomen", title ="Last 10 Women")
plt.show()
```

It is found at that overall, thelast 10 of major ranking, have more share of women 

```python
fig = plt.figure(figsize = (10,10))
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)
recent_grads[:10].plot.bar(ax = ax1 , x = "Major", y = "Unemployment_rate", title = "Unemployment Rate")
ax2 = recent_grads[-10:].plot.bar(ax = ax2 , x = "Major", y ="Unemployment_rate", title ="Unemployment Rate")
plt.show()
```

For the unemployment rate, in the last 10 rank of the major, it has higher unemployment rate compare to the first 10 rank of major
